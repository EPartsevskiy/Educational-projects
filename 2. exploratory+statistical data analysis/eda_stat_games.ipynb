Год выпуска по всем играм с пропущенными названиями - 1993 год. Сумма глобальных продаж - 2.42 млн. Считаем эти данные несущественные и допустимые к удалению.

```python
data = d.dropna(subset=['year_of_release', 'name']).reset_index(drop=True)
data.info()
```

Обработаем **пропущенные значения** в столбце `'rating'`. Для обработки определим: количество пропусков в столбце; количество игр с пропусками в данном столбце, но заполненными значениям по наименованию игр (разные платформы).

```python
empty_rating = data[data['rating'].isna()]
full_rating = data[data['rating'].notna()]
full_rating = full_rating.drop_duplicates(subset='name', keep='first')
display("Количество пропущенных значений в столбце 'rating': {}".
        format(empty_rating['name'].count()))
display("Количество игр с пропущенными значения в столбце 'rating', но с укзанным значением по другой платформе: {}".
       format(empty_rating.query('name in @full_rating.name')['name'].count()))
```

Заполним 417 пустых значений в столбце `'rating'` по идентичным значениям стобца `'name'` и `'year_of_release'`.

```python
data_to_fill = empty_rating.query('name in @full_rating.name')

for i in data_to_fill.index:
    name = data_to_fill.loc[i, 'name']
    row = full_rating[full_rating['name'] == name]
    j = row.index[0]
    data.loc[i, 'rating'] = full_rating.loc[j, 'rating']
```

Остальные пропущенные значения в столбце `'rating'` заменим на значение `'unknown'`.

```python
data['rating'] = data['rating'].fillna('unknown')
```

### Обработка значений `tbd` в столбце `user_score`

Аббревиатура `tbd` расшифровывается как 'to be determined (будет определено)'. Таким образом, можно предположить, что данные значения просто забыли занести позже.

```python
display("Количество строк со значением 'tbd': {}".format(data[data['user_score'] == 'tbd']['user_score'].count()))
display(data[data['user_score'] == 'tbd'].head())
```

В качестве обработки `tbd` значений предлагается внести значения `'critic_score' / 10`.

```python
def tbd_fill(row):
    if row['user_score'] == 'tbd':
        return round(float(row['critic_score']) / 10, 1)
    return row['user_score']

data['user_score_corr'] = data.apply(lambda row: tbd_fill(row), axis=1)
data['user_score_corr'] = data['user_score_corr'].astype(float)
display(data[data['user_score'] == 'tbd'].head())
```

Пропуски в столбцах 'critic_score', 'user_score', 'rating' не будем заполнять по причине того, что значения в данных столбцах невозможно определить или логически обоснвоано задать по имеющимся данным.

### Описание и визуализация продаж по платформам

Определим платформы с максимальным количеством прожаж.

```python
display('Количество платформ в даных: {}'.format(data['platform'].nunique()))
platforms = data.groupby(by='platform')['sum_sales'].sum()
platforms = platforms.sort_values(ascending=False)
pop_plat = platforms[:11]

fig, ax = plt.subplots(figsize=(8,5))

ax.bar(x=pop_plat.index, height=pop_plat.values)
ax.grid()
ax.set_ylabel('Сумма продаж')
ax.set_xlabel('Наименование платформы')
ax.set_title('Продажи на топ {} платформах'.format(pop_plat.count()))
plt.show()
```

**Вывод:** всего в даных **31** различных платформ. Для анализа "жизненного цикла популярности" платформ по года из 31 платформы были отобраны **11** с продажами более 250 млн. по каждой.

Определим **"жизненный цикл популярности"** платформ. 

```python
pop_platform = data.query('platform in @pop_plat.index & year_of_release > 1994')
data_pop_plat = pop_platform.groupby(by=['platform', 'year_of_release'])['sum_sales'].sum()
data_pop_plat = data_pop_plat.unstack(level=0)
data_pop_plat = data_pop_plat.fillna(0)
dict_pop_plat = data_pop_plat.to_dict('list')
indexes = list(data_pop_plat.index)
colors = ['#f00', '#ffa8a8', '#a11313', '#fbff03', '#fa82d2', '#f803fc', 
          '#a92eab', '#001aff', '#6171ff', '#03eeff', '#03ff39']
width = 0.6

fig, ax = plt.subplots(figsize=(15, 6))
bottom = np.zeros(len(indexes))
cs = 0
for platform, sales in dict_pop_plat.items():
    graphic = ax.bar(indexes, sales, width, label=platform, bottom=bottom, color=colors[cs])
    bottom += sales
    cs += 1
    
ax.set_xticks(indexes)
ax.set_title('Сумма продаж по платформам')
ax.set_ylabel('Сумма продаж')
ax.set_xlabel('Год')
ax.legend(title='Платформа')
plt.show()
```

```python
pop_platform_life = (pop_platform.groupby(by='platform').
                     agg({'year_of_release': ['min', 'max']}))
pop_platform_life.columns = ['start_sales', 'finish_sales']
pop_platform_life['lifetime'] = pop_platform_life['finish_sales'] - pop_platform_life['start_sales']
display(pop_platform_life)
display("Средний срок 'жизни' платформ: {} лет".
        format(round(pop_platform_life['lifetime'].mean(), 0)))
```

**Вывод:**
- Средний срок 'жизни' платформ около 10 - 11 лет;

## Проверка гипотез

### Гипотеза о пользовательских рейтингах платформ Xbox One и PC

**Гипотеза:**

H0: Средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу PC.

Н1: Средний пользовательский рейтинг платформы Xbox One не равен среднему пользовательскому рейтингу PC.

```python
xbox_one_year = new_data.query('platform == "XOne"')['year_of_release'].min()
xbox_one = new_data.query('platform == "XOne"')['user_score_corr']
xbox_one = xbox_one.dropna().astype(float)
pc = new_data.query('platform == "PC" & year_of_release >= @xbox_one_year')['user_score_corr']
pc = pc.dropna().astype(float) 

display('Средний пользовательский рейтинг Xbox One: {}'.format(round(xbox_one.mean(), 1)))
display('Средний пользовательский рейтинг PC: {}'.format(round(pc.mean(), 1)))

alpha = 0.05

results = st.ttest_ind(xbox_one, pc)
display(f'p-значение: {results.pvalue}')

if results.pvalue > alpha:
    display('Средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу PC')
else:
    display('Средний пользовательский рейтинг платформы Xbox One не равен среднему пользовательскому рейтингу PC')
```

### Гипотеза о пользовательских рейтингах жанров Action и Sports

**Гипотеза:**

H0: Средний пользовательский рейтинг игр в жанре Action равен среднему пользовательскому рейтингу игр в жанре Sports.

Н1: Средний пользовательский рейтинг игр в жанре Action не равен среднему пользовательскому рейтингу игр в жанре Sports.

```python
action = new_data.query('genre == "Action"')['user_score_corr']
action = action.dropna().astype(float)
sports = new_data.query('genre == "Sports"')['user_score_corr']
sports = sports.dropna().astype(float) 

display('Средний пользовательский рейтинг игр в жанре Action: {}'.format(round(action.mean(), 1)))
display('Средний пользовательский рейтинг игр в жанре Sports: {}'.format(round(sports.mean(), 1)))

alpha = 0.05

results = st.ttest_ind(action, sports)
display(f'p-значение: {results.pvalue}')

if results.pvalue > alpha:
    display('Средний пользовательский рейтинг игр в жанре Action ' +
            'равен среднему пользовательскому рейтингу игр в жанре Sports')
else:
    display('Средний пользовательский рейтинг игр в жанре Action ' +
            'не равен среднему пользовательскому рейтингу игр в жанре Sports')
```

### Выводы по 5 разделу

- Средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу PC;
- Средний пользовательский рейтинг игр в жанре Action не равен (больше) среднему пользовательскому рейтингу игр в жанре Sports.
